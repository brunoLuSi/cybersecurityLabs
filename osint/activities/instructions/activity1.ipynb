## Activity 1: Guided activity with the website search engine

This activity aims to **familiarize yourself with the use of Shodan’s and Censys’ functions**
to search for devices connected to the Internet.

In order to complete this task, first, you need to **create an account** on either Shodan
(https://www.shodan.io) or **Censys** (https://censys.io).

Once you have created an account on either platform, you can start using the **search engine**
**for exposed devices and services on the Internet**, exploring detailed information such as
IP addresses, open ports, software versions, and other security-relevant data.

Answer the questions using either search engine. At the beginning of this section, you must
specify which of the two search engines you are using for this practice, as you cannot switch
between them.

1. Type ”Amazon” and then press the Enter key. How many results did you get?


2. Now, look at the additional information displayed on the left side of the search engine.
Your search results are categorized into different sections. You can click on any of
them to refine your search. How many results, if any, are there for the location Spain?


3. Let’s continue exploring the categories section on both platforms. If you look closely,
you will see that both pages share a section called **”TOP PORTS”**. In each case,
which port has the highest number of results, and what is the count?


4. Now that you have a basic understanding of how the search engine works, let’s use the queries offered by both platforms. Depending on which scanner you are using, enter the following query in the search bar:

    - Censys: host.location.city="Madrid"
    - Shodan: city="Madrid"

    How many results did you get in your scanner?


5. If you check the **”TOP COUNTRIES”** section in the sidebar, you will notice that
even though Spain appears at the top, there are other cities named Madrid in different
parts of the world. To refine our search further, let’s specify the country along with the
city. Use the following queries:

    - Censys: (host.location.city=”Madrid”) and host.location.country=”Spain”
    - Shodan: city:”Madrid” country:”ES”

    How many results did you get in your scanner ?


6. After exploring these examples of how the search engine works, let’s take advantage
of its features and perform a deeper search on a specific IP address to examine the
available information. Enter the following IP in the search bar:

    - Censys i Shodan: ip:”84.88.66.162”

    In **Shodan**, click on the first result that appears. In **Censys**, it should take you directly
to the result page; if not, follow the same step as in Shodan. Once inside, you will
see a lot of information, much of which may be unfamiliar. Let’s focus on the basics.
**Which ports are open on this IP, and what protocol does each one use?**


7. If you look closely, you will see that both port **25 and 587** are open and use the same
protocol. Investigate the possible reason why each port might serve a different function.




8. Have you detected any vulnerabilities in the previous exercise? If so, specify which
vulnerability or vulnerabilities were detected, on which ports, and explain what each
one means in your own words. (For this question, you must use Shodan, as it is not
possible with Censys).

Once you have learned a bit about how queries work and the information contained within
an IP address, I will now provide you with a set of query cases. Based on a given phrase, you
will need to write the correct query. Some of these queries have been taken from examples
found on the internet, where they existed for both search engines or only for one, in which
case the other had to be created or translated. In other cases, the same label values were
searched first, and then a well-known product, number, or name was looked up.

- **IPs in the range 192.167.0.0/16 with the DNS port open**
- **Windows hosts with autonomous systems 13335**
- **Find websites (HTTP) with directory listing enabled**
- **Expired SSL certificates from France (use the country code)**
- **Filter HTTP with status code 200 and running PHP in the backend**
- **Find Apache servers running version 2.4.7**
- **Hacked web servers**
- **Servers in Russia with remote access protocols**
- **Servers from the game ”Minecraft” in the 1.21 version**
- **Android IP webcam servers**
- **Plex media servers**
- **Mikrotik routers exposed**
