**8. Answer from activity 8**

- Yes, there are vulnerabilities. He has detected two and both on port 443. The first is CVE-2015-4000 and the second is CVE-2015-0204.

- The CVE-2015-4000 vulnerability: In general, this vulnerability allowed an attacker to access and manipulate data in compromised TLS connections. At the time, the TLS protocol had a vulnerability known as "Logjam", which could allow an attacker to intercept sensitive information. This occurred when a system did not properly handle an encryption option called DHE_EXPORT. The attacker could intercept this information from a man-in-the-middle attack in order to degrade the connection to a weak 512-bit encryption. If he managed to gain access, the attacker could recover the session key and modify the content of the traffic.

- The vulnerability CVE-2015-0204: OpenSSL clients accepted weak ciphers, called EXPORT-grade, even if they were not requested. This allowed attackers to intercept the connection through a *man-in-the-middle* attack and thus force the server to use this weak cipher. With this type of cipher, the attacker could easily decrypt the communication between the client and the server.
