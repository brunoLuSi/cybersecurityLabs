```python
import shodan
import sys
import json 
import matplotlib.pyplot as plt

api = shodan.Shodan('YOUR_API_KEY')

FACETS = [
    ('port', 5),
    ('isp', 10),
    ('os', 6),
    ('device', 8),
]

FACET_TITLES = {
    'port': 'Top 5 Ports',
    'isp': 'Top 10 Organizations',
    'os': 'Top 6 OS',
    'device': 'Top 8 Devices',
}

try:
    city = input('Enter the city you want to consult: ').strip()
    
    if not city:
        print('Error: No search city was entered.')
        print('Usage: Enter a valid query to perform the search.')
        sys.exit(1)


    query = f'country:ES city:{city}'
    result = api.count(query, facets=FACETS)

    print('Information about Shodan')
    print ('Query: %s' % query)
    print ('Total results: %s\n' % result['total'])

    for facet in result['facets']:
        print(FACET_TITLES[facet])

        values = []
        counts = []
            
        for term in result['facets'][facet]:
            print('%s: %s' % (term['value'], term['count']))
            values.append(str(term['value']))
            counts.append(term['count'])

        print('')

        # Creation of the bar chart depending on the facet
        plt.figure(figsize=(10, 6))
            
        if facet == 'port':
            plt.bar(values, counts)
            plt.xlabel('Port')
            plt.ylabel('Number of Results')
            plt.title(FACET_TITLES[facet])
            
        elif facet == 'isp':
            plt.barh(values, counts)
            plt.xlabel('Count')
            plt.ylabel('ISP')
            plt.title(FACET_TITLES[facet])
            
        elif facet == 'os' or facet == 'device':
            plt.pie(counts, labels=values, autopct='%1.1f%%')
            plt.title(FACET_TITLES[facet])
            
        plt.tight_layout()
        plt.show()
        
except Exception as e:
    print ('Error: %s' % e)
    sys.exit(1)
```
