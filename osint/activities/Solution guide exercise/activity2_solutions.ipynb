## 2. Differences between Shodan and Censys

#### Data Collection

**Shodan**: In its scanning approach, it is focused on identifying devices and services based on specific queries. Shodan indexes devices by searching for banners and metadata exposed by these devices, which provides information about their configurations.

**Censys**: It performs regular scans of the Internet, collecting data on any server and device. It focuses on collecting detailed information about protocols and services, capturing SSL certificates and identifying vulnerabilities.

#### Data Types

**Shodan**: It is more focused on indexing devices such as webcams, routers, etc. It often captures metadata such as banners that provide information about the type of device and the services it offers.

**Censys**: Censys relies heavily on its data structures about SSL/TLS services and configurations. It offers fairly detailed information about their vulnerabilities and protocol versions.

#### User Interface and API

**Shodan**: Focused on being more casual and accessible to new users. It is basically simple and effective.

**Censys**: Provides detailed analysis tools and also an API to be integrated. It focuses on presenting information clearly.

#### Use Cases

**Shodan**: Used by researchers and professionals related to cybersecurity to assess vulnerabilities, check regulatory compliance and discover devices. These groups prefer this tool as it provides more emphasis on their data and analysis related to cybersecurity.

**Censys**: It is focused on the most extensive use, from researchers as well as individuals who are interested in exploring IoT devices and their hardware. Its wide variety of devices provides more diverse features for making queries.

#### Traffic in the world

**Shodan**: It has more than a million visits per month.

**Censys**: It has more than 400 thousand visits per month.

#### Filter system

**Shodan**: Sometimes, its filtering system falls short compared to Censys.

**Censys**: It has a very good filter system that allows a large number of functions to customize searches.

#### Support and community

In **Shodan** there is more participation thanks to the forums and discussions by the community than in Censys.

#### Scanning transparency

**Censys** publishes the list of its scanning IP lists (centralized in the US) which facilitates the process for system administrators to choose not to be scanned, while in
**Shodan** it does not reveal these IPs, the only thing it reveals is that its scanners use different geographical locations, which allows it to avoid blockages.

#### Data freshness

**Shodan**: Slightly faster than Censys. Updates 20% of data in less than 5 hours.

**Censys**: About 65% of data is updated every 15 hours. 100% in less than 24 hours.

#### Services and Ports Scanned

**Shodan**: More focused on HTTP and IoT devices. Scans a wider set of ports, including uncommon ports.

**Censys**: More focused on HTTPS and administrative services like SSH and FTP. Its port scanning is more thorough for fewer ports.

#### Resource consumption

**Shodan**: Performs fewer SYN scans per day. Its TCP session is longer (2s on average)

**Censys**: Performs more SYN scans per day (411 vs. 176 for Shodan). Its TCP session is shorter (139ms on average)

#### Scan Type

**Shodan**: Its scanning methodology is less transparent. It seems to divide the IPb4 space into regions assigned to different scanners.

**Censys**: Performs horizontal scans: scans the entire IPv4 address space for a single port at a time. Collects service banners once it identifies an open port
